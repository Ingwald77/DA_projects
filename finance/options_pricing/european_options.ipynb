```python
from scipy.stats import norm
import math
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
import arch
import datetime as dt
from pandas_datareader import data as pdr
```

```python
# Delta: measures the sensitivity of the option price to changes in the underlying stock price.
# Gamma: measures the sensitivity of the option DELTA to changes in the underlying stock price.
# Vega: measures the sensitivity of the option price to changes in the stock price volatility.
# Theta: measures the sensitivity of the option price to the passage of time.
# Rho: measures the sensitivity of the option price to changes in the risk-free interest rate.

# Delta is very useful, because it is the number of shares to buy (or sell) to hedge out 
# the risk of changes in the underlying stock’s price when short a call (or put) option.
# In other words, if you have a portfolio short 1 call option and long Delta shares of stock, 
# then your portfolio is riskless (over very short time periods). 
# This is referred to as delta hedging.
# Similarly, a portfolio short one put and short Detla shares of stock is riskless.
# Call Deltas range from 0 to 1, and put Deltas range from -1 to 0.

def bsm_delta(S, X, T, r, sigma, option_type):
    d1 = (math.log(S/X) + (r + 0.5*sigma**2) * T) /sigma * math.sqrt(T)
    if option_type == 'C':
        return norm.cdf(d1)
    if option_type == 'P':
        return norm.cdf(d1) - 1
    
# The higher the Gamma (in absolute value) the more often you’ll need to rebalance 
# a delta-neutral portfolio.
# Suppose the Gamma of a call option on a stock is 0.03.
# This means that a $1 increase in the stock’s price will increase the Delta of the option by 0.03.
# To create a Gamma-neutral portfolio, you’ll have to trade in an option on the 
# underlying stock – or some derivative which is not linearly related to the underlying stock.
    
def bsm_gamma(S, X, T, r, sigma):
    d1 = (math.log(S/X) + (r + 0.5*sigma**2) * T) /sigma * math.sqrt(T)
    return norm.pdf(d1) / (S * sigma * math.sqrt(T))

# The higher Vega is in absolute value, the more sensitive the option’s price is to changes in volatility.
# The Vega of a long position in a call or put option is positive.
# The Vega of an at-the-money option decreases as expiration approaches.
# If the Vega of a put option is 40, then a 1% increase in volatility increases 
# the value of the option by 1%×40=$0.40.
# Often Vega is simply quoted as the change in the option’s price for each 1% change 
# in volatility (the 0.40 above).

def bsm_vega(S, X, T, r, sigma):
    d1 = (np.log(S/X) + (r + 0.5*sigma**2)*T) / sigma * np.sqrt(T)
    return S * norm.pdf(d1) * np.sqrt(T) 
    
# Theta is usually negative, which means as time passes the option’s premium declines 
# (all other variables held constant).
# Theta is calculated in years, but if we divide theta by 252, we get the daily decline 
# in the option premium solely due to time decay.
# For example, say Theta is -25, then in days Theta is −25252=−0.09921
# which means all else constant, the option’s price will decline by $0.09921 per day.
# Theta is often quoted as this time decay per day (-0.9921 above).

def bsm_theta(S, X, T, r, sigma, option_type):
    d1 = (math.log(S/X) + (r + 0.5*sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    if option_type == 'C':
        return -(S * norm.pdf(d1) * sigma) / (2 * math.sqrt(T)) - (r * X * math.exp(-r * T) * norm.cdf(d2))
    elif option_type == 'P':
        return -(S * norm.pdf(d1) * sigma) / (2 * math.sqrt(T)) + (r * X * math.exp(-r * T) * norm.cdf(-d2))
    
# Rho measures the sensitivity of the option value to interest rates.
# For example, say for a put option Rho is -30.
# Then, if the risk-free rate increases by 1%, the option’s value will decline by 1%×30=$0.30
# Often Rho is simply quoted as the change in the option’s price for each 1% change in 
# the risk-free rate (the 0.30 above).

def bsm_rho(S, X, T, r, sigma, option_type):
    d1 = (math.log(S/X) + (r + 0.5*sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    if option_type == 'C':
        return X * T * math.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'P':
        return -X * T * math.exp(-r * T) * norm.cdf(-d2)
```

```python
# d1 represents the excess return of the underlying stock (i.e., the stock's 
# expected return minus the risk-free rate) divided by the volatility of 
# the stock's returns. d2 represents the excess return of the stock minus the present value 
# of the strike price, also divided by the stock's volatility.
# S - current price of underlying asset
# X - option strike price of underlying asset
# T - time to expiration in years
# r - ontiniuosly compounded risk-free-rate
# sigma - volatility of underlying asset

def bsm_pricing(S, X, T, r, sigma, option_type):
    d1 = (np.log(S/X) + (r + 0.5*sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'C':
        return S * norm.cdf(d1) - np.exp(-r * T) * X * norm.cdf(d2)
    else:
        return  np.exp(-r * T) * X * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
# Historical volatility is calculated by taking the standard deviation 
# of the log returns over a given time period

def hist_volatility(data, window=30):
#     returns = data.pct_change()[1:]
#     returns = price_data/price_data.shift(1)
#     log_returns = np.log(price_data/price_data.shift(1))
    log_returns = np.log(price_data) - np.log(price_data.shift(1))
    vol = log_returns.rolling(window).std() * np.sqrt(252)
    return vol

ticker = 'TSLA'
start_date = '2018-01-01'
end_date = '2023-02-14'
price_data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']

hist_vol = hist_volatility(price_data, 30)

fig, ax1 = plt.subplots()
ax1.plot(price_data, color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price')
ax2 = ax1.twinx()
ax2.plot(hist_vol, color='red')
ax2.set_ylabel('Volatility %')
plt.title('{} price and historical volatility'.format(ticker))
plt.show()
```

```python
def implied_volatility_iter(S, X, T, r, market_price, option_type):
    
    def option_price_diff(sigma, S, X, T, r, market_price, option_type):
        model_price = bsm_pricing(S, X, T, r, sigma, option_type)
        return abs(model_price - market_price)

    
    best_vol = None
    best_price_diff = float('inf')
    price_diff = None

    sigma_range = np.arange(0.01, 2, 0.01)

    for sigma in sigma_range:

        price_diff = option_price_diff(sigma, S, X, T, r, market_price, option_type)
        if price_diff < best_price_diff:
            best_price_diff = price_diff
            best_vol = sigma
            
    return best_vol

print(implied_volatility_iter(100, 110, 1, 0.05, 10.5, 'C'))
```

```python
def implied_volatility_nr(S, X, T, r, market_price, option_type):
    max_iterations = 1000
    tolerance = 0.00001
    sigma = 0.5
    
    for i in range(max_iterations):
        ### calculate difference between blackscholes price and market price with
        diff = market_price - bsm_pricing(S, X, T, r, sigma, 'C')

        if abs(diff) < tolerance:
            return sigma

        ### use newton rapshon to update the estimate of sigma
        vega = bsm_vega(S, X, T, r, sigma)
        sigma = sigma + diff / vega
        
    return sigma

print(implied_volatility_nr(100, 110, 1, 0.05, 10.5, 'C'))
```

```python
# buying one At-The-Money (ATM) call option and selling the Out-Of-The-Money call option

# Define the underlying asset's price range
spot_price = np.arange(50, 150, 1)

# Define the strike prices for the long and short call options
long_call_strike_price = 100
short_call_strike_price = 110

# Calculate the payoff for the long call option
long_call_payoff = np.maximum(spot_price - long_call_strike_price, 0)

# Calculate the payoff for the short call option
short_call_payoff = np.maximum(short_call_strike_price - spot_price, 0)

# Calculate the payoff for the bull call spread
bull_call_spread_payoff = long_call_payoff - short_call_payoff

# Plot the results
plt.plot(spot_price, long_call_payoff, label='Long Call')
plt.plot(spot_price, short_call_payoff, label='Short Call')
plt.plot(spot_price, bull_call_spread_payoff, label='Bull Call Spread')
plt.legend()
plt.xlabel('Underlying Asset Price')
plt.ylabel('Payoff')
plt.title('Bull Call Spread Payoff Diagram')
plt.show()
```

```python
# simultaneously purchase a call and put option on the same underlying asset with 
# the same strike price and expiration date

# Define the underlying asset's price range
spot_price = np.arange(50, 150, 1)

# Define the strike price for the call and put options
strike_price = 100

# Calculate the payoff for the call option
call_payoff = np.maximum(spot_price - strike_price, 0)

# Calculate the payoff for the put option
put_payoff = np.maximum(strike_price - spot_price, 0)

# Calculate the payoff for the long straddle
long_straddle_payoff = call_payoff + put_payoff

# Plot the results
plt.plot(spot_price, call_payoff, label='Call Option')
plt.plot(spot_price, put_payoff, label='Put Option')
plt.plot(spot_price, long_straddle_payoff, label='Long Straddle')
plt.legend()
plt.xlabel('Underlying Asset Price')
plt.ylabel('Payoff')
plt.title('Long Straddle Payoff Diagram')
plt.show()
```
