### Столбцы данных для проведения исследования
Набор данных, используемых для проведения исследования содержит следующие столбцы:
- country — страна происхождения вина.<br>
- description — описание вкусового профиля вина.<br>
- designation — виноградник, ягоды которого используются при изготовлении вина.<br>
- points — оценка вина, согласно рейтинга журнала Wine Enthusiast, в баллах по шкале от 1 до 100.<br>
- price — стоимость одной бутылки вина.<br>
- province — место изготовления вина (провинция или штат).<br>
- region_1 — винодельческая область в провинции или штате (например, долина Напа в штате Калифорния).<br>
- region_2 — (необязательно) конкретный зона виноделия в винодельческой области (например, Резерфорд в долине Напа).<br>
- variety — сорт винограда, который используется при изготовлении вина (например, Пино Нуар).<br>
- winery — винодельня-производитель вина.<br>

```python
# Определить популярные сорта вина в бюджетном сегменте по сорту вина
# взято значение цены меньеш 24, что меньше медианного
df_budget = df.copy()
def budget_rating(row):
    if row['price'] < 24 and row['points'] >= 90:
        val = 1
    else:
        val = 0
    return(val)
df_budget['budget'] = df_budget.apply(budget_rating, axis=1)
df_budget[df_budget['budget'] == 1].sort_values(['variety', 'points'],ascending=[True,False]).sample(5)
```

```python
# уникальные винодельни, которые продают бюджетное вино
i.groupby('country')['winery'].nunique().sort_values(ascending=False).head(3)
```

```python
# Построить график «ящик с усами» по рейтингам, выполнив категоризацию по странам
fig = plt.figure(figsize=(18,12))
top_countries = df[df['points'] == 100]['country'].unique()
points_top_reviewed_wines = df[df.variety.isin(top_varieties)]
x = points_top_reviewed_wines.country
y = points_top_reviewed_wines.points
sns.boxplot(x,y)
plt.xlabel('Country', fontsize=20)
plt.ylabel('Ratings', fontsize=20)
plt.title('Top countries by rating distributions',fontsize=20);
plt.xticks(rotation='vertical')
plt.figtext(0.5, -0.1, "Рисунок 2.3 - Топ стран по распределнию рейтинга вина",fontsize=15);
```

В рамках исследования статистических показателей необходимо построить линейную регрессию зависимости между ценой продукта и его рейтингом.

```python
from yellowbrick.regressor import PredictionError, ResidualsPlot
from sklearn.linear_model import LinearRegression
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.api as sm
```

```python
df.price.describe() # медиана(24) и среднее значение(32,8) очень сильно различаются
```

```python
plt.figure(figsize = (15,5))
sns.boxplot(df.price,width=0.8).set_title('Visualisation of the amount and placing of raw price outliers');
plt.figtext(0.5, -0.1, "Рисунок 4.1 - Распределения выбросов по цене",fontsize=15);
```

```python
Q1 = df.price.quantile(0.25)
Q3 = df.price.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df.price < (Q1 - 1.5 * IQR)) | (df.price > (Q3 + 1.5 * IQR)))] # отсечём выбросы
```

```python
plt.figure(figsize = (15,5))
sns.boxplot(df.price,width=0.8).set_title('Visualisation of the amount and placing of processed price outliers');
plt.figtext(0.5, -0.1, "Рисунок 4.2 - Рапсределения выбросов по цене после отсечения выбросов",fontsize=15);
```

```python
df.shape
```

```python
# разобьём данные на тренировочный и тестовый набор данных
x = df[['points']].astype('float')
y = df.price
a = df.points # понадобится для построения scatterplot


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
# построим модель линейной регрессии
linreg = LinearRegression()
linreg.fit(x_train, y_train)
print (linreg.intercept_)
print (linreg.coef_)
```

```python
y_line_pred = linreg.predict(x_test)
```

```python
#здесь я использую библиотеку statsmodels потому что мне нравится описание статистики
x_const_sm = sm.add_constant(x_train)
lr = sm.OLS(y_train, x_const_sm).fit()
lr.summary()
```

```python
print ('MAE:', mean_absolute_error(y_line_pred,x_test))
print ('MSE:', mean_squared_error(y_line_pred,x_test))
print ('RMSE:', np.sqrt(mean_squared_error(y_line_pred,x_test)))
```

```python
plt.figure(figsize=(24,8))
visualizer = PredictionError(linreg)
visualizer.fit(x_train, y_train);
# Fit the training data to the visualizer
visualizer.score(x_test, y_test);
# Evaluate the model on the test data
plt.figtext(0.5, -0.1, "Рисунок 4.3 - Предсказание цены по странам",fontsize=20);
visualizer.poof();
```

```python
# график показывает разницу между остаточными значениями на вертикальной оси и зависимой переменной на горизонатльной оси
plt.figure(figsize=(24,8))
visualizer = ResidualsPlot(linreg) 
visualizer.fit(x_train,y_train); 
visualizer.score(x_test,y_test);
plt.figtext(0.5, -0.1, "Рисунок 4.4 - Остаточные значения при построении модели обучения",fontsize=20);
visualizer.poof();
```
